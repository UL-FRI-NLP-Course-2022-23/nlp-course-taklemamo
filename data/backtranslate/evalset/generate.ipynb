```python
import numpy as np
import pandas as pd

df = pd.read_csv("../testset/T5-large_results.csv", sep="\t", index_col=0)
df
```

```python
np.random.seed(42)
df = df[df.inputs.apply(len) < 200]
idxs = np.random.permutation(len(df))[:10]
df.iloc[idxs]
```

```python
idxs
```

```python
ref_path = "../testset/backtranslate.csv"
paths = {
    "baseline": "../testset/baseline_results.csv",
    "T5-small": "../testset/T5-small_results.csv",
    "T5-large": "../testset/T5-large_results.csv"
}

out = {
    "paragraph": [],
    "paraphrase": [],
    "source": [],
    "id": [],
}
for name, path in paths.items():
    data = pd.read_csv(path, sep="\t", index_col=0)
    data = data.iloc[idxs]
    out["paragraph"] += data.inputs.to_list()
    out["paraphrase"] += data.outputs.to_list()
    out["source"] += [name] * len(data)
    out["id"] += list(idxs)
```

```python
out["paragraph"]
```

```python
data = pd.read_csv(ref_path, sep="\t", names=["inputs", "outputs"])
data = data.iloc[idxs]
out["paragraph"] += data.inputs.to_list()
out["paraphrase"] += data.outputs.to_list()
out["source"] += ["reference"] * len(data)
out["id"] += list(idxs)
```

```python
out_df = pd.DataFrame(out)
out_df
```

```python
np.random.seed(1337)
out_df = out_df.iloc[np.random.permutation(len(out_df))].reset_index(drop=True)
```

```python
out_df.to_csv("eval.csv", sep="\t")
```
